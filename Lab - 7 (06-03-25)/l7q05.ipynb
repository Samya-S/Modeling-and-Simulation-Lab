```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 20
dt = 0.01
t = np.linspace(0, T, int(T/dt) + 1)
```

```python
x0 = 1
y0 = 0
```

```python
def u(x, y):
    return -y

def v(x, y):
    return x
```

```python
# Estimate by Runge-Kutta Method
x_rk = np.zeros(len(t))
x_rk[0] = x0

y_rk = np.zeros(len(t))
y_rk[0] = y0

for i in range(1, len(t)):    
    xk1 = u(x_rk[i-1], y_rk[i-1])
    yk1 = v(x_rk[i-1], y_rk[i-1])
    xk2 = u(x_rk[i-1] + xk1 * dt/2, y_rk[i-1] + yk1 * dt/2)
    yk2 = v(x_rk[i-1] + xk1 * dt/2, y_rk[i-1] + yk1 * dt/2)
    xk3 = u(x_rk[i-1] + xk2 * dt/2, y_rk[i-1] + yk2 * dt/2)
    yk3 = v(x_rk[i-1] + xk2 * dt/2, y_rk[i-1] + yk2 * dt/2)
    xk4 = u(x_rk[i-1] + xk3 * dt, y_rk[i-1] + yk3 * dt)
    yk4 = v(x_rk[i-1] + xk3 * dt, y_rk[i-1] + yk3 * dt)

    x_rk[i] = x_rk[i-1] + (1/6) * dt * (xk1 + 2*xk2 + 2*xk3 + xk4)
    y_rk[i] = y_rk[i-1] + (1/6) * dt * (yk1 + 2*yk2 + 2*yk3 + yk4)
```

```python
r = np.sqrt(x_rk**2 + y_rk**2)
```

```python
plt.plot(t, x_rk, label='x(t)', color='blue')
plt.plot(t, y_rk, label='y(t)', color='red')
plt.plot(t, r, label='r(t)', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Time-Dependent Flow')
plt.legend()
plt.grid()

plt.show()
```

```python
plt.plot(x_rk, y_rk, label='Path', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Pathlines in Rotational Flow')
plt.legend()
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```
