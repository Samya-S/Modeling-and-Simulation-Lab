```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def potential_function(x, y):
    r = np.sqrt(x**2 + y**2) + 1e-10
    return np.log(r)

def velocity_field(x, y):
    r = np.sqrt(x**2 + y**2) + 1e-10
    u = x / r**2
    v = y / r**2
    return u, v
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

Phi = potential_function(X, Y)
U, V = velocity_field(X, Y)
```

```python
plt.figure()
plt.streamplot(X, Y, U, V, density=1)
plt.contour(X, Y, Phi, levels=20, cmap='viridis', alpha=0.5)
plt.title('Irrotational Source Flow')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.colorbar(label='Potential Function ϕ = ln(r)')
plt.show()
```
