```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = 1
v = 0
```

```python
x0 = 0
y0 = 0
```

```python
t = np.linspace(0, 100)
x = t
y = np.zeros(len(t))
```

```python
r = np.zeros(len(t))

for i in range (0, len(t)):
    r[i] = np.sqrt(x[i]**2 + y[i]**2)
```

```python
# plt.plot(t, y, label='y(t)', color='blue')
# plt.plot(t, x, label='x(t)', color='red')
plt.plot(t, r, label='Path', color='black')
plt.scatter([x0], [y0], color='red', label='Starting Point (0,0)')
plt.xlabel('Time')
plt.ylabel('Pathline')
plt.title('Pathline for Uniform Flow')
plt.legend()
plt.grid()

plt.show()
```

```python
plt.plot(x, y, label='Path', color='black')
plt.scatter([x0], [y0], color='red', label='Starting Point (0,0)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Pathline for Uniform Flow')
plt.legend()
plt.grid()

plt.show()
```
