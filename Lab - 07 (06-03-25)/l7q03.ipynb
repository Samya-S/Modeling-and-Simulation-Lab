```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating data set
w = 10
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = -Y
V = X
 
# Creating plot
fig = plt.figure()
plt.streamplot(X, Y, U, V, density = 1)
plt.title("Streamlines for a Vortex")
plt.gca().set_aspect('equal', adjustable='box')

# show plot
plt.show()
```
