```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def source_sink_flow(X, Y):
    U = (X + 1) / ((X + 1)**2 + Y**2) - (X - 1) / ((X - 1)**2 + Y**2)
    V = Y / ((X + 1)**2 + Y**2) - Y / ((X - 1)**2 + Y**2)
    return U, V
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

U, V = source_sink_flow(X, Y)
```

```python
plt.figure()
plt.streamplot(X, Y, U, V, density=1)
plt.title('Streamlines for Combined Flow due to Source and Sink')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter([-1], [0], color='red', s=100, label='Source')
plt.scatter([1], [0], color='green', s=100, label='Sink')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
