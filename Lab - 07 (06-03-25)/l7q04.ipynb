```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating data set
w = 10
Y, X = np.mgrid[0:w:100j, -w:w:100j]
U = Y
V = X*0
 
# Creating plot
fig = plt.figure()
plt.streamplot(X, Y, U, V, density = [0.1, 1])
plt.title("Flow Over a Flat Plate")

# show plot
plt.show()
```
