```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = 5
m = 100
c = 2
```

```python
x = 3  
y = 4

z = x + 1j * y
```

```python
# Flow Past a Rankine Oval
def F(z):
    return U*z + (m/(2*np.pi))*np.log(z + c) + (m/(2*np.pi))*np.log(z - c)
```

```python
def phi(z):
    return np.real(F(z))

def psi(z):
    return np.imag(F(z))
```

```python
print(f"Flow Past a Rankine Oval at ({x}, {y}): phi = {phi(z)}, psi = {psi(z)}")
```

```python
x_range = np.linspace(-3, 3, 100)
y_range = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x_range, y_range)

phi_values = np.zeros_like(X)
psi_values = np.zeros_like(X)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        phi_values[i, j] = psi(X[i, j] + 1j*Y[i, j])
        psi_values[i, j] = phi(X[i, j] + 1j*Y[i, j])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.contour(X, Y, psi_values, levels=20, cmap='viridis')
plt.colorbar(label="Stream Function")
plt.title(f"Streamlines (U={U}, m={m})")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.contour(X, Y, phi_values, levels=20, cmap='plasma')
plt.colorbar(label="Velocity Potential")
plt.title(f"Equipotential Lines (U={U}, m={m})")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)

plt.show()
```
