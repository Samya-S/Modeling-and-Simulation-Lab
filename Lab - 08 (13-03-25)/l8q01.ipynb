```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define the velocity field components

```python
def u(x, y):
    return 2*x + y

def v(x, y):
    return -x - 3*y
```

### Define a function to compute partial derivatives numerically using finite differences


```python
def partial_derivative(f, x, y, var, h=1e-5):
    if var == 'x':
        return (f(x + h, y) - f(x - h, y)) / (2 * h)
    elif var == 'y':
        return (f(x, y + h) - f(x, y - h)) / (2 * h)
```

### Calculate the partial derivatives at (x, y) = (1, 2)

```python
x, y = 1, 2

du_dx = partial_derivative(u, x, y, 'x')
dv_dy = partial_derivative(v, x, y, 'y')
```

```python
# Sum of the partial derivatives (should be zero for incompressibility)
continuity_value = du_dx + dv_dy
```

```python
# Print the results
print(f"∂u/∂x at (1, 2): {du_dx}")
print(f"∂v/∂y at (1, 2): {dv_dy}")
print(f"Final sum: {continuity_value}")
```

```python
# Verify the continuity equation (it should be 0 for incompressible flow)
if np.isclose(continuity_value, 0):
    print("Continuity equation verified: ∇ · v = 0")
else:
    print("Continuity equation not satisfied.")
```

### Visualization of the velocity field divergence (EXTRA)

```python
# Create grid
x_vals = np.linspace(-5, 5, 20)
y_vals = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x_vals, y_vals)
```

```python
# Compute velocity components
U = u(X, Y)
V = v(X, Y)
```

```python
# Calculate the divergence
divergence = np.gradient(U, axis=1) + np.gradient(V, axis=0)
```

```python
# Plot the divergence
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, divergence, levels=20, cmap='coolwarm')
plt.colorbar(label='Divergence')
plt.title('Divergence of the Velocity Field')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
