```python
import numpy as np
import matplotlib.pyplot as plt
```

### Velocity components


```python
def u(x, y):
    return y

def v(x, y):
    return -x
```

### Runge-Kutta method to solve the system of ODEs


```python
def runge_kutta(x0, y0, t, dt):
    x = np.zeros(len(t))
    y = np.zeros(len(t))
    x[0] = x0
    y[0] = y0
    
    for i in range(1, len(t)):
        k1x = u(x[i-1], y[i-1]) * dt
        k1y = v(x[i-1], y[i-1]) * dt
        
        k2x = u(x[i-1] + 0.5 * k1x, y[i-1] + 0.5 * k1y) * dt
        k2y = v(x[i-1] + 0.5 * k1x, y[i-1] + 0.5 * k1y) * dt
        
        k3x = u(x[i-1] + 0.5 * k2x, y[i-1] + 0.5 * k2y) * dt
        k3y = v(x[i-1] + 0.5 * k2x, y[i-1] + 0.5 * k2y) * dt
        
        k4x = u(x[i-1] + k3x, y[i-1] + k3y) * dt
        k4y = v(x[i-1] + k3x, y[i-1] + k3y) * dt
        
        x[i] = x[i-1] + (1/6) * (k1x + 2*k2x + 2*k3x + k4x)
        y[i] = y[i-1] + (1/6) * (k1y + 2*k2y + 2*k3y + k4y)
    
    return x, y
```

```python
# Initial conditions
x0, y0 = 1, 0

# Time steps
T = 5
dt = 0.01
t = np.linspace(0, T, int(T/dt) + 1)

# Compute the pathline using Runge-Kutta method
x_vals, y_vals = runge_kutta(x0, y0, t, dt)
```

### Plot the pathline


```python
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Pathline')
plt.scatter(x0, y0, color="red", label="Starting point")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Pathline for the given velocity field')
plt.grid(True)
plt.legend()
plt.show()
```
