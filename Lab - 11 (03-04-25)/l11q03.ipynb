```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.2
b = 0.01
c = 0.1
d = 0.02
x0 = 30
y0 = 10
```

```python
dt = 0.1
T = 200
t = np.linspace(0, T, int(T/dt) + 1)
```

```python
def f(x, y):
    dxdt = a*x - b*x*y
    dydt = -c*y + d*x*y
    return dxdt, dydt
```

```python
x = np.zeros(len(t))
y = np.zeros(len(t))
x[0] = x0
y[0] = y0
```

```python
for i in range(1, len(t)):
    k1x, k1y = f(x[i-1], y[i-1])
    k2x, k2y = f(x[i-1] + k1x * dt / 2, y[i-1] + k1y * dt / 2)
    k3x, k3y = f(x[i-1] + k2x * dt / 2, y[i-1] + k2y * dt / 2)
    k4x, k4y = f(x[i-1] + k3x * dt, y[i-1] + k3y * dt)
    
    x[i] = x[i-1] + (dt / 6) * (k1x + 2*k2x + 2*k3x + k4x)
    y[i] = y[i-1] + (dt / 6) * (k1y + 2*k2y + 2*k3y + k4y)
```

```python
plt.plot(t, x, label='Prey (x)', color='blue')
plt.plot(t, y, label='Predator (y)', color='red')
plt.xlabel('Time (in days)')
plt.ylabel('Population')
plt.title('Predator-Prey System')
plt.legend()
plt.grid()
plt.show()
```
