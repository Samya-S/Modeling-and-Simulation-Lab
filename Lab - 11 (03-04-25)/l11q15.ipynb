```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 0.1
K = 500
P0 = 50
```

```python
dt = 0.1
T = 200
t = np.linspace(0, T, int(T/dt) + 1)
```

```python
def f(P, h):
    return r * P * (1 - P / K) - h
```

```python
h_critical = r * K / 4
P = np.zeros(len(t))
P[0] = P0
```

```python
for i in range(1, len(t)):
    k1 = f(P[i-1], h_critical)
    k2 = f(P[i-1] + k1 * dt / 2, h_critical)
    k3 = f(P[i-1] + k2 * dt / 2, h_critical)
    k4 = f(P[i-1] + k3 * dt, h_critical)
    P[i] = P[i-1] + (dt / 6) * (k1 + 2*k2 + 2*k3 + k4)
    if P[i] < 0:
        P[i] = 0
        break
```

```python
plt.plot(t, P, label='Population (P)', color='blue')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Logistic Growth Model with Critical Harvesting')
plt.legend()
plt.grid()
plt.show()
```
