```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rk2(U, h, N):
    dy = h / (N - 1)
    y = np.linspace(0, h, N)
    u = np.zeros(N)
    
    for i in range(1, N):
        k1 = (U / h)
        k2 = (U / h)
        u[i] = u[i - 1] + dy * (k1 + k2) / 2
    
    return y, u
```

```python
U = 0.5
h = 0.01
N = 50
```

```python
y, u = rk2(U, h, N)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(u, y, label='RK Solution')
plt.xlabel('Velocity u (m/s)')
plt.ylabel('Height y (m)')
plt.title('RK Solution of Couette Flow')
plt.legend()
plt.grid()
plt.show()
```
