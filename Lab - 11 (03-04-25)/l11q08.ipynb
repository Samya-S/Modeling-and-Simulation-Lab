```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = 0.01
dPdx = -100
mu = 1.0
```

```python
y = np.linspace(0, H, 100)
```

```python
u = (1 / (2 * mu)) * dPdx * (y**2 - H * y)
```

```python
plt.plot(u, y)
plt.xlabel('Velocity u (m/s)')
plt.ylabel('y (m)')
plt.title('Velocity Profile Between Parallel Plates')
plt.grid()
plt.show()
```
