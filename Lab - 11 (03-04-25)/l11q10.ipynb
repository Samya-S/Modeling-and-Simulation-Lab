```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def stream_function(x, y, Q, x_s, y_s):
    return (Q / (2 * np.pi)) * np.arctan2((y - y_s), (x - x_s))
```

```python
Q = 10
source_pos = (-2, 0)
sink_pos = (2, 0)
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

```python
psi = (stream_function(X, Y, Q, *source_pos) - stream_function(X, Y, Q, *sink_pos))
```

```python
plt.figure(figsize=(8, 6))
plt.contour(X, Y, psi, levels=50, cmap='jet')
plt.scatter([source_pos[0], sink_pos[0]], [source_pos[1], sink_pos[1]], color='red', s=100, marker='o')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Streamlines of Source-Sink Pair')
plt.show()
```
