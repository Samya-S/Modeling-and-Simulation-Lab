```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = 2
R = 1
```

```python
x = np.linspace(-3*R, 3*R, 100)
y = np.linspace(-3*R, 3*R, 100)
X, Y = np.meshgrid(x, y)
```

```python
Theta = np.arctan2(Y, X)
R_grid = np.sqrt(X**2 + Y**2)

Vr = U * (1 - (R**2 / R_grid**2)) * np.cos(Theta)
Vtheta = -U * (1 + (R**2 / R_grid**2)) * np.sin(Theta)

Vx = Vr * np.cos(Theta) - Vtheta * np.sin(Theta)
Vy = Vr * np.sin(Theta) + Vtheta * np.cos(Theta)

Vx[R_grid < R] = 0
Vy[R_grid < R] = 0
```

```python
plt.figure(figsize=(6, 6))
plt.streamplot(X, Y, Vx, Vy)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Streamlines of Flow Around a Cylinder')
circle = plt.Circle((0, 0), R, color='black', fill=True)
plt.gca().add_patch(circle)
plt.axis('equal')
plt.grid()
plt.show()
```
