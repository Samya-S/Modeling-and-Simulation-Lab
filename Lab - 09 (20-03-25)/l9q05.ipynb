```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
mu = 1  # Strength of the doublet
r_min, r_max = 0.1, 2 
num_points = 100  
```

```python
# Create a grid of points in polar coordinates
x = np.linspace(-r_max, r_max, num_points)
y = np.linspace(-r_max, r_max, num_points)
X, Y = np.meshgrid(x, y)
```

```python
# Calculate the radial distance (r) from the origin and angle (theta)
R = np.sqrt(X**2 + Y**2) 
Theta = np.arctan2(Y, X)  
```

```python
# Avoid division by zero at the origin
R[R == 0] = np.nan
```

```python
# Stream function: ψ = -μ sin(θ) / (2πr)
psi = -mu * np.sin(Theta) / (2 * np.pi * R)

# Velocity potential: φ = -μ cos(θ) / (2πr)
phi = -mu * np.cos(Theta) / (2 * np.pi * R)
```

```python
# Plotting the streamlines (contour of stream function)
plt.figure(figsize=(12, 6))

# Streamlines
plt.subplot(1, 2, 1)
plt.contour(X, Y, psi, 250, cmap='jet') 
plt.title("Streamlines for Flow Around a Doublet")
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label="Stream function (ψ)")
plt.grid(True)
plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')

# Equipotential lines
plt.subplot(1, 2, 2)
plt.contour(X, Y, phi, 250, cmap='viridis') 
plt.title("Equipotential Lines for Flow Around a Doublet")
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label="Velocity potential (ϕ)")
plt.grid(True)
plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```
