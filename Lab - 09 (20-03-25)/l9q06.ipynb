```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants
U = 1.0  
a = 1.0  
```

```python
# Define grid for polar coordinates (r, theta)
r_min = 0.5 
r_max = 3.0 
theta_min = 0  
theta_max = 2 * np.pi  
```

```python
# Create meshgrid for r and theta
r = np.linspace(r_min, r_max, 100)
theta = np.linspace(theta_min, theta_max, 100)
R, Theta = np.meshgrid(r, theta)
```

```python
# Convert polar to cartesian coordinates for plotting
X = R * np.cos(Theta)
Y = R * np.sin(Theta)
```

```python
# Compute the stream function Ïˆ
psi = U * R * np.sin(Theta) * (1 - (a**2 / R**2))
```

```python
# Compute the radial and tangential velocities
v_r = U * np.cos(Theta) * (1 - (a**2 / R**2))  
v_theta = -U * np.sin(Theta) * (1 / R + a**2 / R**3)  
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(6, 6))

# Plot streamlines
ax.contour(X, Y, psi, 20, cmap='inferno')

# Plot the cylinder (circle)
cylinder = plt.Circle((0, 0), a, color='black', linewidth=2, fill=False)
ax.add_artist(cylinder)

# Add labels and settings
ax.set_xlim([-r_max, r_max])
ax.set_ylim([-r_max, r_max])
ax.set_aspect('equal')
ax.set_title('Flow Around a Cylinder')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Show the plot
plt.show()
```
