```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1  # Strength of the source and sink
a = 1  # Distance of source/sink from the origin
```

```python
# Create a grid of points (x, y)
x = np.linspace(-2, 2, 100)  
y = np.linspace(-2, 2, 100)  
X, Y = np.meshgrid(x, y) 
```

```python
# Compute r1 and r2 
r1 = np.sqrt((X - a)**2 + Y**2)  # Distance from the source at (a, 0)
r2 = np.sqrt((X + a)**2 + Y**2)  # Distance from the sink at (-a, 0)
```

```python
# Compute the stream function psi
psi = (m / (2 * np.pi)) * np.log(r1 / r2)
```

```python
# Plot the streamlines using contour plot
plt.figure()
plt.contour(X, Y, psi, levels=np.linspace(-2, 2, 20), cmap='jet')  
plt.colorbar(label='Stream Function (ψ)') 
plt.title("Streamlines for a Source and Sink Pair")
plt.xlabel("X")
plt.ylabel("Y")
plt.axhline(0, color='black',linewidth=1)  
plt.axvline(0, color='black',linewidth=1)  
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
