```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
U = 1
x_min, x_max = 0, 10 
y_min, y_max = -5, 5 
num_points = 100
```

```python
# Create a grid of points in Cartesian coordinates
x = np.linspace(x_min, x_max, num_points)
y = np.linspace(y_min, y_max, num_points)
X, Y = np.meshgrid(x, y)
```

```python
# Velocity potential for uniform flow: φ = U * x
phi = U * X
```

```python
# Plotting the equipotential lines using contour
plt.figure(figsize=(8, 6))
plt.contour(X, Y, phi, 20, cmap='jet')
plt.title("Equipotential Lines for Uniform Flow along the X-Axis")
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label="Velocity potential (ϕ)")
plt.grid(True)
plt.axis('equal')
plt.show()
```
