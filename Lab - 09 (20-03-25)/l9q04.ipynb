```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
m = 1  # Sink strength
r_min, r_max = 0.1, 5
num_points = 100
```

```python
# Create a grid of points in Cartesian coordinates
x = np.linspace(-r_max, r_max, num_points)
y = np.linspace(-r_max, r_max, num_points)
X, Y = np.meshgrid(x, y)
```

```python
# Calculate the radial distance (r) from the origin and angle (theta)
R = np.sqrt(X**2 + Y**2) 
Theta = np.arctan2(Y, X) 
```

```python
# Avoid division by zero at the origin
R[R == 0] = np.nan
```

```python
# Calculate the radial velocity components (v_r = -m / (2 * pi * r))
v_r = -m / (2 * np.pi * R) 
v_x = v_r * (X / R) 
v_y = v_r * (Y / R) 
```

```python
# Plotting the streamlines using streamplot
plt.figure(figsize=(8, 6))
plt.streamplot(X, Y, v_x, v_y, linewidth=1)
plt.title("Streamlines of Flow Due to a Sink at the Origin")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
