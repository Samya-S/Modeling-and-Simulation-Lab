```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = 1.0  # Uniform flow velocity
m = 1.0  # Source strength
```

```python
# Define grid for polar coordinates (r, theta)
r_min = 0.1 
r_max = 3.0  
theta_min = 0 
theta_max = 2 * np.pi  
```

```python
# Create meshgrid for r and theta
r = np.linspace(r_min, r_max, 100)
theta = np.linspace(theta_min, theta_max, 100)
R, Theta = np.meshgrid(r, theta)
```

```python
# Convert polar to cartesian coordinates for plotting
X = R * np.cos(Theta)
Y = R * np.sin(Theta)
```

```python
# Compute the stream function Ïˆ
psi = U * Y + (m / (2 * np.pi)) * Theta
```

```python
# Create the plot
fig, ax = plt.subplots()

# Plot streamlines
ax.contour(X, Y, psi, 20, cmap='inferno')

# Add labels and settings
ax.set_xlim([-r_max, r_max])
ax.set_ylim([-r_max, r_max])
ax.set_aspect('equal')
ax.set_title('Combined Source and Uniform Flow')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Show the plot
plt.show()
```
