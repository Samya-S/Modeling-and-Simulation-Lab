```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
U = 1
x_min, x_max = 0, 10
y_min, y_max = -5, 5
num_points = 100  
```

```python
# Create a grid of points in Cartesian coordinates
x = np.linspace(x_min, x_max, num_points)
y = np.linspace(y_min, y_max, num_points)
X, Y = np.meshgrid(x, y)
```

```python
# Velocity components for uniform flow: U in x-direction, 0 in y-direction
v_x = U * np.ones_like(X)
v_y = np.zeros_like(Y)
```

```python
# Plotting the streamlines using streamplot
plt.figure(figsize=(8, 6))
plt.streamplot(X, Y, v_x, v_y, linewidth=1)
plt.title("Streamlines of Uniform Flow along the X-Axis")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
