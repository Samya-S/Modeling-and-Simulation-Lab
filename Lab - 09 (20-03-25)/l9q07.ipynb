```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Strength of the stagnation flow
a = 1.0  
```

```python
# Define grid for x and y coordinates
x_min = -2.0  
x_max = 2.0   
y_min = -2.0  
y_max = 2.0   
```

```python
# Create meshgrid for x and y
x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Compute the stream function Ïˆ
psi = a * (X**2 - Y**2)
```

```python
# Create the plot
fig, ax = plt.subplots()

# Plot streamlines
ax.contour(X, Y, psi, 20, cmap='inferno')

# Add labels and settings
ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_aspect('equal')
ax.set_title('Flow Near a Stagnation Point')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Show the plot
plt.show()
```
