```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the system of ODEs using Euler's method
def euler_method(velocity_field, initial_conditions, t_span, dt, num_steps):
    x_vals = []
    y_vals = []
    t_vals = np.linspace(t_span[0], t_span[1], num_steps)
    
    for (x0, y0) in initial_conditions:
        x = x0
        y = y0
        
        x_vals.append([x])
        y_vals.append([y])

        for t in t_vals[1:]:
            dxdt = x
            dydt = -y
            x = x + dxdt * dt
            y = y + dydt * dt

            x_vals[-1].append(x)
            y_vals[-1].append(y)

    return t_vals, x_vals, y_vals
```

```python
# Initial conditions for fluid particles
initial_conditions = [
    (1, 1),
    (2, 1),
    (-1, -1),
    (1, -2),
    (-2, 2)
]
```

```python
# Time span and parameters
t_span = (0, 5)  
dt = 0.01  
num_steps = int((t_span[1] - t_span[0]) / dt)
```

```python
# Solve the system using Euler's method
t_vals, x_vals, y_vals = euler_method(None, initial_conditions, t_span, dt, num_steps)
```

```python
# Create the plot
plt.figure()

# Plot the pathlines
for i in range(len(initial_conditions)):
    plt.plot(x_vals[i], y_vals[i], label=f'Initial condition: ({initial_conditions[i][0]}, {initial_conditions[i][1]})')

# Add labels and settings
plt.title('Pathlines in a Simple Flow (u = x, v = -y) using Euler\'s Method')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
